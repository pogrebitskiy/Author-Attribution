```python
import seaborn as sns
import pickle
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('../data/cleaned_articles.pkl')
df.head()
```

```python
# sns plot of number of articles per author
plt.figure(figsize=(10, 6))
sns.barplot(x=df['author'].value_counts().index, y=df['author'].value_counts().values)
plt.xticks(rotation=90);
plt.title('Number of articles per author')
plt.xlabel('Author')
plt.ylabel('Number of articles')
plt.show()
```
